```
import pandas as pd
from kaggle_evaluation.aimo_3_gateway import AIMO3Gateway
from fractions import Fraction
import math

def solve_algebra(expr):
    expr = expr.replace('=', '-(') + ')'
    for x in range(-2000, 2001):
        try:
            if eval(expr.replace('x', str(x))) == 0:
                return x
        except:
            pass
    return 0

def solve_number_theory(text):
    nums = [int(s) for s in text.replace(',', ' ').split() if s.lstrip('-').isdigit()]
    if 'gcd' in text and len(nums) >= 2:
        return math.gcd(nums[0], nums[1])
    if 'lcm' in text and len(nums) >= 2:
        return nums[0] * nums[1] // math.gcd(nums[0], nums[1])
    return 0

def solve_geometry(text):
    nums = [Fraction(s) for s in text.replace('/', ' ').split() if s.replace('-', '').isdigit()]
    if 'triangle' in text and len(nums) >= 2:
        return nums[0] * nums[1] / 2
    return 0

def dss_omega_solver(problem: str):
    p = problem.lower()
    if 'x' in p and '=' in p:
        return solve_algebra(problem)
    if any(k in p for k in ['gcd', 'lcm']):
        return solve_number_theory(p)
    if 'triangle' in p:
        return solve_geometry(p)
    return 0

def predict(batch: pd.DataFrame) -> pd.DataFrame:
    return pd.DataFrame({
        'id': batch['id'],
        'answer': [dss_omega_solver(p) for p in batch['problem']]
    })

AIMO3Gateway(predict).run()
```
